```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import time

base_url = "https://www.analyticsvidhya.com/courses/"

# Store courses
courses = []


for page in tqdm(range(1, 11)):
    if page == 1:
        url = base_url
    else:
        url = f"{base_url}?page={page}"

    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch page {page}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")

    # Each course is inside div with class 'card-body'
    for course_div in soup.select("div.card-body"):

        title_tag = course_div.select_one("h2.card-text")
        title = title_tag.get_text(strip=True) if title_tag else None


        duration_tag = course_div.select_one("p.fs-14 span")
        duration = duration_tag.get_text(strip=True) if duration_tag else None


        rating_tag = course_div.select_one("div.float-end")
        rating = rating_tag.get_text(strip=True) if rating_tag else None


        learners_tag = course_div.select_one("button.no-pointer")
        learners = learners_tag.get_text(strip=True) if learners_tag else None

        # Analytics Vidhya is mostly Free or Paid, here we assume Free if not specified
        price = "Free"  # default

        # Difficulty (not available on page HTML, set None)
        difficulty = None

        instructor = "Analytics Vidhya"

        # Category / Domain
        category = "AI / ML / Data Science"

        # Platform
        platform = "Analytics Vidhya"

        courses.append({
            "Title": title,
            "Platform": platform,
            "Category": category,
            "Price": price,
            "Duration": duration,
            "Difficulty": difficulty,
            "Instructor": instructor,
            "Rating": rating,
            "Learners / Reviews": learners
        })

    # Optional delay to be polite
    time.sleep(1)

# Save to CSV
df = pd.DataFrame(courses)
df.to_csv("analyticsvidhya_courses.csv", index=False)
print(f"✅ Saved {len(df)} courses")
```
